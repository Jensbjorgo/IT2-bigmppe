```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 3*x**2
```

```python
x=[]
y=[]

for i in range(20):
    x.append(i)
    y.append(f(i))
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,200,3)
y=f(x)
y=y*2
```

```python
plt.plot(x,y)
plt.title("$f(x)=3x^2$")
plt.style.use("classic")
plt.xlabel("antall dyr")
plt.ylabel("Mengden bsj kg")
plt.show()
```

```python
xverdier = np.linspace(0, 20, 50)

# Graf 1
yverdier = 2*xverdier+1 

plt.subplot(2, 2, 1)
plt.plot(xverdier, yverdier)
plt.grid()

# Graf 2
yverdier = xverdier**2-3 

plt.subplot(2, 2, 2)
plt.plot(xverdier, yverdier)
plt.grid()

plt.show()

```
